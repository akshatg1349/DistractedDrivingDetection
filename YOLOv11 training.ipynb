```python
import shutil
import os

def create_directory(directory_path):
    # Check if the directory exists
    if not os.path.exists(directory_path):
        # If it doesn't exist, create it
        os.makedirs(directory_path)
        print(f"Directory '{directory_path}' created.")
    else:
        print(f"Directory '{directory_path}' already exists.")
```

```python
# Specify the directory path you want to check/create
model_name = f'yolo11_best_model_epoch_{EPOCHS}_lr_{LEARNING_RATE}'
model_folder = '/content/drive/MyDrive/1:1_Akshat_Gupta/Models'
destination_path = os.path.join(model_folder, model_name)

# Call the function to check and create the directory
create_directory(destination_path)

# IMPORTANT
# define the best model custom data path
source_path = '/content/runs/detect/yolo11s_custom'
shutil.move(source_path, destination_path)
```
